## Total cost 

```python
#define a cost calculating function that returns the latency of an edge
#based on Youn et al
#t(x)=t0[1+α(x/c)^β]
def latency(r_csv, edge, edge_removed):
    
    flow = 0
    
    #if the edge removed is a string
    if type(edge_removed) == str:
        flow = r_csv.at[edge,edge_removed]
    else: #if the edge removed is  a number   
        edge_removed_string = str(edge_removed) + "_removed"
        #If the flow is NAN (the edge was removed), return 0
        flow = r_csv.at[edge,edge_removed_string]
    
    if math.isnan(flow) == True:
        return 0
    
    alpha = 0.15
    beta = 4
    t0 = r_csv.at[edge,"free flow time"]
    capacity = r_csv.at[edge,"capacity"]
    
    return t0*(1 + alpha*(flow/capacity)**beta)

# returns the total cost for equilibrium with a certain edge removed
def total_cost(r_csv, edge_removed):
    
    ttl_cost = 0
    for i in range(results_csv.shape[0]):
        ttl_cost = ttl_cost + latency(r_csv,i,edge_removed)
    
    return ttl_cost

#returns dataframe with total cost of flow for each link removed
def total_cost_all(r_csv):
    temp_df = pd.DataFrame(index = results_csv.index, columns = ["ttl_cost_when_removed"]) 
    for i in range(results_csv.shape[0]):
        temp_df.at[i,"ttl_cost_when_removed"] = total_cost(r_csv,i)
    
    return temp_df
    
```

# Compute node centrality measures

Centrality measures to compute are available here https://networkx.org/documentation/stable/reference/algorithms/centrality.html

```python
#Here we compute measures for all the nodes in the Souix Falls graph
sfalls_graph = SiouxFalls_dict['SiouxFalls']['None']['fw_run'].graph

degree_centralities = nx.degree_centrality(sfalls_graph)
in_degree_centralities = nx.in_degree_centrality(sfalls_graph)
out_degree_centralities = nx.out_degree_centrality(sfalls_graph)
eigen_centralities = nx.eigenvector_centrality(sfalls_graph)
betweenness_centralities = nx.betweenness_centrality(sfalls_graph)
```

```python
#Now we want to add these to the dataframe
sfalls_reg_df = temp_df

sfalls_reg_df['o_degree_centrality'] = 0
sfalls_reg_df["d_degree_centrality"] = 0
sfalls_reg_df['int_degree_centrality'] = 0
sfalls_reg_df['o_in_degree_centrality'] = 0
sfalls_reg_df['d_in_degree_centrality'] = 0
sfalls_reg_df['int_in_degree_centrality'] = 0
sfalls_reg_df['o_out_degree_centrality'] = 0
sfalls_reg_df['d_out_degree_centrality'] = 0
sfalls_reg_df['int_out_degree_centrality'] = 0
sfalls_reg_df['o_eigen_centrality'] = 0
sfalls_reg_df['d_eigen_centrality'] = 0
sfalls_reg_df['int_eigen_centrality'] = 0
sfalls_reg_df['o_bw_centrality'] = 0
sfalls_reg_df['d_bw_centrality'] = 0
sfalls_reg_df['int_bw_centrality'] = 0

for i in range(len(sfalls_reg_df)):
    sfalls_reg_df.loc[i,'o_degree_centrality'] = degree_centralities[sfalls_reg_df.at[i,'init node']]
    sfalls_reg_df.loc[i,'d_degree_centrality'] = degree_centralities[sfalls_reg_df.at[i,'term node']]
    sfalls_reg_df.loc[i,'int_degree_centrality'] = sfalls_reg_df.loc[i,'o_degree_centrality']*sfalls_reg_df.loc[i,'d_degree_centrality']
    sfalls_reg_df.loc[i,'o_in_degree_centrality'] = in_degree_centralities[sfalls_reg_df.at[i,'init node']]
    sfalls_reg_df.loc[i,'d_in_degree_centrality'] = in_degree_centralities[sfalls_reg_df.at[i,'term node']]
    sfalls_reg_df.loc[i,'int_in_degree_centrality'] = sfalls_reg_df.loc[i,'o_in_degree_centrality']*sfalls_reg_df.loc[i,'d_in_degree_centrality']
    sfalls_reg_df.loc[i,'o_out_degree_centrality'] = out_degree_centralities[sfalls_reg_df.at[i,'init node']]
    sfalls_reg_df.loc[i,'d_out_degree_centrality'] = out_degree_centralities[sfalls_reg_df.at[i,'term node']]
    sfalls_reg_df.loc[i,'int_out_degree_centrality'] = sfalls_reg_df.loc[i,'o_out_degree_centrality']*sfalls_reg_df.loc[i,'d_out_degree_centrality']
    sfalls_reg_df.loc[i,'o_eigen_centrality'] = eigen_centralities[sfalls_reg_df.at[i,'init node']]
    sfalls_reg_df.loc[i,'d_eigen_centrality'] = eigen_centralities[sfalls_reg_df.at[i,'term node']]
    sfalls_reg_df.loc[i,'int_eigen_centrality'] = sfalls_reg_df.loc[i,'o_eigen_centrality']*sfalls_reg_df.loc[i,'d_eigen_centrality']
    sfalls_reg_df.loc[i,'o_bw_centrality'] = betweenness_centralities[sfalls_reg_df.at[i,'init node']]
    sfalls_reg_df.loc[i,'d_bw_centrality'] = betweenness_centralities[sfalls_reg_df.at[i,'term node']]
    sfalls_reg_df.loc[i,'int_bw_centrality'] = sfalls_reg_df.loc[i,'o_bw_centrality']*sfalls_reg_df.loc[i,'d_bw_centrality']

sfalls_reg_df['log_capacity'] = np.log(sfalls_reg_df['capacity'].astype('float64'))
sfalls_reg_df['log_eq_flow'] = np.log(sfalls_reg_df['eq_flow'].astype('float64'))
    
sfalls_reg_df['ttl_cwr_log_change'] = sfalls_reg_df['ttl_cost_when_removed'] / none_eq_cost
sfalls_reg_df['ttl_cwr_log_change'] = np.log(sfalls_reg_df['ttl_cwr_log_change'].astype('float64'))

#ttl_cwr_log_change = sfalls_reg_df['ttl_cost_when_removed'] / none_eq_cost
#ttl_cwr_log_change = np.log(ttl_cwr_log_change.astype('float64'))

    
sfalls_reg_df
```

# Run linear regressions

```python
#Run linear regression

y = sfalls_reg_df['ttl_cwr_log_change']
X = sfalls_reg_df.loc[:, 'o_in_degree_centrality':'log_eq_flow']

# adding the constant term
X = sm.add_constant(X)
 
# performing the regression
# and fitting the model
result = sm.OLS(y, X).fit(cov_type='HC1')
 
# printing the summary table
print(result.summary())
print(result.summary().as_latex())
```

```python
X_small = sfalls_reg_df.loc[:, 'd_bw_centrality']

# adding the constant term
X_small = sm.add_constant(X_small)
 
# performing the regression
# and fitting the model
result = sm.OLS(y, X_small).fit(cov_type='HC1')
 
# printing the summary table
print(result.summary())
print(result.summary().as_latex())
```
